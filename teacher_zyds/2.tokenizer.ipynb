```python
from transformers import AutoTokenizer
```

```python
sen = '彻底疯狂！求求你阻止我！'
sen
```

```python
tokenizer = AutoTokenizer.from_pretrained('uer/roberta-base-finetuned-dianping-chinese')
```

```python
tokenizer
```

```python
tokenizer.save_pretrained('./tokenizer')
```

```python
tokenizer = AutoTokenizer.from_pretrained('./tokenizer/')
tokenizer
```

```python
#第二步开始分词
token = tokenizer.tokenize(sen)
token
```

```python
#查看词典
tokenizer.vocab
```

```python
tokenizer.vocab_size
```

```python
#开始转换   从 原始字符到数字
ids = tokenizer.convert_tokens_to_ids(token)
ids
```

```python
token = tokenizer.convert_ids_to_tokens(ids)
token
```

```python
strings = tokenizer.convert_tokens_to_string(token)
strings
```

```python
#下面展示更加简便通用的一次性输入的方法
tokens = tokenizer.encode(sen)
print(tokens) 
tokens = tokenizer.encode(sen,add_special_tokens=True)
print(tokens) 
```

```python
sen1 = tokenizer.decode(tokens)
print(sen1)
sen1 = tokenizer.decode(tokens,skip_special_tokens=False)
print(sen1)
```

```python
#填充和截断
tokens = tokenizer.encode(sen,padding='max_length',max_length=2,truncation=True)
tokens
```

```python
#attention mask 和token id
attn_mask = [1 if idx!=0 else 0 for idx in ids]
attn_mask,ids
```

```python
#都是第一个句子
token_type_ids = [0] * len(ids)
token_type_ids
```

```python
#简便方法把上诉功能都实现了  其实直接用tokenizer就是
tokens = tokenizer.encode_plus(sen,padding='max_length',max_length=20,truncation=True)
tokens
```

```python
tokenizer(sen,padding='max_length',max_length=20,truncation=True)
```
